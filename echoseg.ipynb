```python
from ultralytics import YOLO
import cv2
import numpy as np

# Choose a segmentation model â€“ try nano first for CPU speed
model = YOLO("yolov8n-seg.pt")   # or "yolov8s-seg.pt" if CPU is decent
video_path = r"C:\Users\sadiq\echo4.mp4"  # put your MP4 next to the notebook or give full path
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    raise RuntimeError(f"Cannot open video: {video_path}")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Optional: downscale for CPU speed
    h, w = frame.shape[:2]
    scale = 5   # 50% size; tune as needed
    frame_small = cv2.resize(frame, (int(w*scale), int(h*scale)))

    # YOLOv8-Seg inference (CPU by default if no GPU)
    results = model(frame_small, verbose=False)  # list with one result

    annotated = results[0].plot()  # draws masks + boxes on the frame

    cv2.imshow("YOLOv8-Seg Echo", annotated)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
